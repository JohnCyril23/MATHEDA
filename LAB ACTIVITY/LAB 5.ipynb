```python
#imports and packages

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
df = pd.read_csv(r'C:\Users\luxah\OneDrive\MATHEDA\LAB ACTIVITY\DataSet\current-power.csv',
                 delimiter=",")

df
```

```python
#summary of statistics

df.describe()
```

```python
df.cov()
```

```python
#scatter plot

x = df['Current']
y = df['Power']

plt.Figure()
plt.scatter(x,y)
plt.title('Device Current vs Power Response')
plt.xlabel('Current (A)')
plt.ylabel('Power (W) ')
plt.show
```

```python
#Correlation

df.corr()
```

```python
#Regression Line
slope, intercept, r_value, p_value, std_err = stats.linregress(x,y)
line = slope*x + intercept

#Scatter Plot
plt.figure(figsize= (7,7))

plt.scatter(x, y,
    color = "#D9FF00",
    alpha= 0.6,
    label = 'Data Points'
)
plt.plot(x, line,
         color = "#170483",
          label = f'Regression Line (r = {r_value:.3f})' )

plt.title('Device Currnent vs Power Response')
plt.xlabel('Current(A)')
plt.ylabel('Power (W)')
plt.legend()
plt.show()
```

```python
# Air Quality

air = pd.read_csv(r'C:\Users\luxah\OneDrive\MATHEDA\LAB ACTIVITY\DataSet\air-quality.csv', 
                 delimiter=",")

air
```

```python
x = air['NOx']
y = air["Relative_Humidity"]


#Regression Line
slope, intercept, r_value, p_value, std_err = stats.linregress(x,y)
line = slope * x + intercept

#Scatter Plot
plt.figure(figsize= (7,7))

plt.scatter(x, y,
    color = "#D9FF00",
    alpha= 0.6,
    label = 'Data Points'
)
plt.plot(x, line,
         color = "#170483",
          label = f'Regression Line (r = {r_value:.3f})' )

plt.title('Device Currnent vs Power Response')
plt.xlabel('NOx')
plt.ylabel('Humidity %')
plt.legend()
plt.show()
```
